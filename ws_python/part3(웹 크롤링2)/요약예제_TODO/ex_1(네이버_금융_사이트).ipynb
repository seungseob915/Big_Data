```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome('driver/chromedriver')    # 크롬을 사용하기 위해서 chrome driver를 사용해야함
driver.get("https://finance.naver.com")    # 창을 연다
```

```python
xpath="""//*[@id="content"]/div[1]/div[1]/div[1]/div/ul/li[1]/span/a"""
driver.find_element_by_xpath(xpath).click()
```

```python
# 본문을 slicing 하기
import sys
from urllib.request import urlopen

f=urlopen(driver.current_url)    # 현재 드라이버 url 확인
encoding = f.info().get_content_charset(failobj="utf-8")    # 인코딩 방식 확인

text= f.read().decode(encoding)    # 추출한 인코딩 방식으로 다시 디코딩
# print(text)
startidx = text.find('<div class="articleCont" id="content">')
endidx = text.find("<a href", startidx) 

startidx =  startidx + len('<div class="articleCont" id="content">' )
#print(text[ startidx :endidx ]   )

temp = (text[ startidx :endidx ]   )

from bs4 import BeautifulSoup
soup=BeautifulSoup(temp)

total=''
for i in soup:
    total+=i.get_text()
#total.remove('\t','')
```

```python
# 문제2. 워드 클라우드 만들기
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

spwords=set(STOPWORDS) ## 기본적으로 많이 쓰는 영문 단어 ( ex : is)

# 워드 클라우드를 설정합니다. 한글이 있기 때문에 폰트도 넣어 줍니다.(워드 클라우드에는 한글폰트가 없음)
## 핵심
wordcloud = WordCloud(max_font_size=200,font_path='c:/Windows/fonts/malgun.ttf',    
                      stopwords=spwords, ## 기본적으로 많이 쓰는 단어는 제외(예외처리)
                      background_color='#FFFFFF' , random_state=100,
                      colormap="spring",
                      width=1200,height=800).generate(total)

#사이즈 등을 설정해 줍니다.
plt.figure(figsize=(10,8))
plt.imshow(wordcloud)         # 시각화
plt.tight_layout(pad=0)
plt.axis('off')

plt.savefig('wordcloud_image1.png', bbox_inches='tight') # 파일로 저장
plt.show()
```
