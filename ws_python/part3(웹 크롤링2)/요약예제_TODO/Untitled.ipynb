```python
from selenium import webdriver
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
# 리스트로 일단 기사 url 추출 뒤 for문 형식으로 기사 크롤링
driver = webdriver.Chrome('driver/chromedriver')
driver.get('https://finance.naver.com/')
first = urlopen(driver.current_url)
encoding = first.info().get_content_charset(failobj='utf-8')
bsf = BeautifulSoup(first.read().decode(encoding))
Lists = bsf.find('div',class_='section_strategy')
News_List = Lists.find_all('li')
count =1
for title in News_List:
    print('\t\t---------------------------------------{}번 뉴스------------------------------------------------'.format(count))
    count = count+1
    news = re.search(r'<a href=(.*?)onclick',str(title)).group(1)
    news = re.sub(r'\"','',news)
    news = re.sub(r'amp;','',news)
    news = 'https://finance.naver.com' + news
    f = urlopen(news)
    encoding = f.info().get_content_charset(failobj = 'utf-8')
    text = f.read().decode(encoding)
    bs = BeautifulSoup(text)
    text = (bs.find('div',class_="articleCont"))
    test = re.compile('.*?<a href.*?>')
    text = test.search(str(text)).group(0)
    body = re.sub('<.*?>','',str(text))
    print(body.strip())
```
