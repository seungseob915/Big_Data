```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
text=open("test.txt","r", encoding="utf-8").read()    #기존 파일에서 데이터를 읽어옴
```

```python
spwords=set(STOPWORDS) ## 기본적으로 많이 쓰는 단어는 제외
spwords.add('박승섭') ##  제외하고 싶은 단어 추가
spwords.add('박찬호') ##  제외하고 싶은 단어 추가
spwords.add('박근원') ##  제외하고 싶은 단어 추가
spwords.add('카카오톡') ##  제외하고 싶은 단어 추가
spwords.add('오후') ##  제외하고 싶은 단어 추가
spwords.add('오전') ##  제외하고 싶은 단어 추가


# 워드 클라우드를 설정합니다. 한글이 있기 때문에 폰트도 넣어 줍니다.
wordcloud = WordCloud(max_font_size=200,font_path='c:/Windows/fonts/malgun.ttf',
                      stopwords=spwords,background_color='#FFFFFF',
                      width=1200,height=800).generate(text)

plt.figure(figsize=(10,8))
plt.imshow(wordcloud)
plt.tight_layout(pad=0)
plt.axis('off')

plt.savefig('test1.png', bbox_inches='tight')
plt.show()
```
