```python
from urllib.request import urlopen     # urllib.request 에서 urlopen만 임포트 / f=urllib.request.urlopen('http://hanbit.co.kr') 과 동일하다
```

```python
f = urlopen('http://naver.co.kr')
```

```python
# urlopen() 함수는 HTTPResponse 자료형의 객체를 반환
# 파일 객체이므로 open() 함수로 반환되는 파일 객체처럼 핸들링
type(f)    #데이터 타입: HTTPResponse
```

```python
# read() 메서드로 HTTP 응답 본문(bytes 자료형)을 추출
# HTTP 연결은 자동으로 close 되므로 별도의 close() 함수를 호출
f.read()
```
