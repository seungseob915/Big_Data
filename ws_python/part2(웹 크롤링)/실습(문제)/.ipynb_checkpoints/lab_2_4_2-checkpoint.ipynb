```python
naver_movie_url = "https://movie.naver.com/movie/point/af/list.nhn"
```

```python
# 문제 : naver_movie_url 변수의 url을  요청 
from urllib.request import urlopen

uo=urlopen(naver_movie_url)
bytes_content=uo.read() 
```

```python
# 문제 : HTTP 헤더를 기반으로 인코딩 방식을 추출
scanned_text = bytes_content[:1024].decode('ascii', errors='replace')

import re
import sys
match=re.search('charset=["\']?([\w-]+)', scanned_text)
if match:
    encoding = match.group(1)
else:
    encoding='utf-8'

print('encoding:', encoding, file=sys.stderr)
```

```python
# 문제 : 추출한 인코딩 방식으로 디코딩하여,  응답 받은 데이터 출력  
text=bytes_content.decode(encoding)
print(text)
 
```

```python
# 문제 : 위에서 추출한 text에서 '한글' 찾아서 출력 (정규표현식 활용)
re_test=re.compile('[가-힣]+')
re_test.findall(text)
```
